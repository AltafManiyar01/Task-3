### Top 10 States by Sales and Profits 

```python
Top_10_sales = df.groupby("State")["Sales"].sum().nlargest(n=10)
Top_10_sales
```

```python
Top_10_profit = df.groupby("State")["Profit"].sum().nlargest(n=10)
Top_10_profit
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#sns.s
Top_10_sales.plot(kind='bar', figsize=(14, 6))
plt.xlabel("States", fontsize=13)
plt.ylabel("Total Sales", fontsize=13)
plt.title('Top 10 States by Sales', fontsize=16)
plt.show()
```

```python


# Assuming 'State' contains state names and 'Profit' contains profit values
sns.barplot(x='State', y='Profit', data=Top_10_profit.reset_index(), palette='magma')
plt.xlabel("States", fontsize=13)
plt.ylabel("Total Profit", fontsize=13)
plt.title("Top 10 States by Profit", fontsize=16)
plt.xticks(rotation=45)  # Rotate x-axis labels if necessary
plt.tight_layout()  # Adjust layout for better visualization
plt.show()
```

```python
# Assuming 'df' contains the DataFrame with columns 'Sales', 'Profit', and 'Discount'
plt.figure(figsize=(15, 8))
scatter = plt.scatter(x='Sales', y='Profit', c='Discount', s=20, marker='*', data=df, cmap='viridis')
plt.colorbar(scatter, label='Discount')  # Add color bar for 'Discount'

plt.ylabel("Total Profits", fontsize=13)
plt.title("Interdependence of Sales, Profit, and Discount", fontsize=16)
plt.show()
```

```python
financial = df[["Sales","Quantity","Profit","Discount"]]
sns.pairplot(financial)
plt.show()
```

### In above we see that there is some relation between sales and profit and also there is some relation between Discount and Profit. Now To see what exact relation between those entities we plot the heat_map. so we get more clearity

```python
correlation = financial.corr()
sns.heatmap(correlation, xticklabels = correlation.columns, yticklabels =correlation.columns, annot = True)
```

#### From Above map we infer that, 1)sales and profits are positively correlated 2)Discount and Profits are Negatively correlated
